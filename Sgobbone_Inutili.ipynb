```python
conteggio = df['RainTomorrow'].value_counts().to_frame('cont').reset_index()
fig = px.bar(conteggio,
             x = 'RainTomorrow',
             y = 'cont',
             title = 'Analisi di sbilanciamento del target value di training',
             color = 'RainTomorrow')
fig.show()
```

```python
def barre(x, color):
    cont = df[[x,color]].groupby([color]).value_counts().to_frame('conteggio').reset_index()
    fig = px.bar(cont, x = x, y = 'conteggio', color = color, barmode = 'group')
    fig.show()

def scatole(x, color):
  conteggio = df[x].value_counts().to_frame('cont').reset_index()
  fig = px.box(conteggio, x = x, title= x)
  fig.show()

for i in categoriche:
  barre(i,'RainTomorrow')
  scatole(i,'RainTomorrow')


```
