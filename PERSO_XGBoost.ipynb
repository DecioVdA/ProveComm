#### Analisi del bilanciamento del dataset

```python
conteggio = df['Channel'].value_counts().to_frame('cont').reset_index()
fig = px.bar(conteggio, x = 'Channel', y = 'cont', color = 'Channel', title = 'AnalisiBilanciamento ')
fig.show()
```

```python
def istogrammi(x):
  fig = px.histogram(df, x = x, color='Channel', nbins=20, histnorm='percent', title = f'Analisi di {x}', barmode='group')
  fig.show()

def scatole(x):
  fig = px.box(df, x = x, color = 'Channel', title = f'Box plot di {x}')
  fig.show()

categorie = list(set(df.columns)-{'Channel'})

for i in df[categorie]:
  istogrammi(i)
  scatole(i)
```

#### Matrice di correlazione

```python
figura = px.imshow(df.corr(numeric_only=True),
                   text_auto='.2f',
                   range_color= [-1,1],
                   height=700,
                   width = 700,
                   title = 'Analisi della Correlazione tra le variabili numeriche')
figura.show()
```
